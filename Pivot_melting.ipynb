```python
import pandas as pd
import seaborn as sns

titanic_df=sns.load_dataset('titanic')
titanic_df.head()
```

```python
# show the average age and fare of passengers for each class and gender combination.
pivot_table = titanic_df.pivot_table(values=['age', 'fare'], index='pclass', columns='sex', aggfunc='mean')
```
